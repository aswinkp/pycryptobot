```python
from collections import namedtuple
from datetime import datetime
import pandas as pd
from IPython.display import display, HTML


Kline = namedtuple('Kline',["open_time", "open", "high", "low", "close", "volume", "close_time", "quote_asset_vol", "number_of_trades", "base_volume", "quote_volume", "ignore"])
kline_array = []
for i, kline in enumerate(klines, start=1):
    kline_array.append(Kline(*kline))
    
percentage_change = 1
data = []
last_kline = None

next_1hr_positives = 0
next_3hr_positives = 0
next_1hr_negatives = 0
next_3hr_negatives = 0

for i, kline in enumerate(kline_array, start=1):
    if i == 1:
        last_kline=kline
        continue

    last_1hr = round((float(kline_array[i-4].close) - float(kline.close))/float(last_kline.close)*100, 2)
    last_15_min = round((float(kline.close) - float(last_kline.close))/float(last_kline.close)*100, 2)
    try:
        next_1hr = round((float(kline_array[i+4].close) - float(kline.close))/float(last_kline.close)*100, 2)
        next_3hr = round((float(kline_array[i+12].close) - float(kline.close))/float(last_kline.close)*100, 2)
        if next_1hr>0:
            next_1hr_positives+=1
        else:
            next_1hr_negatives+=1
        if next_3hr>0:
            next_3hr_positives+=1
        else:
            next_3hr_negatives+=1
    except IndexError:
        next_1hr = 0
        next_3hr = 0
        
    
    
    if(last_15_min>1):
        data.append([last_1hr, last_15_min, next_1hr, next_3hr])

    last_kline=kline
df = pd.DataFrame(data, columns=["Last 1 hr", "Last 15 min", "Next 1 hr", "Next 3 hr"])
display(HTML(df.to_html()))
```
